```python
h = RECONSTRUCTORS[ len(RECONSTRUCTORS)-1 ]._current_hypergraph
h
```

```python
# transform for visualisation
h1 = dict()
i  = 0
for h_edge in h:
    for j in range(h[h_edge]):
        h1[i] = tuple(h_edge)
        i += 1

H1 = hnx.Hypergraph(h1)
```

```python
# transform for visualisation
h2 = dict()
i  = 0
largest = 1
for h_edge in h:
    el = len(h_edge)
    if el > largest:
        largest = el
for h_edge in h:
    if len(h_edge) == largest:
        for j in range(h[h_edge]):
            h2[i] = tuple(h_edge)
            i += 1

H2 = hnx.Hypergraph(h2)
```

```python
# transform for visualisation
h3 = dict()
i  = 0
for h_edge in h:
    if len(h_edge) > 2:
        for j in range(h[h_edge]):
            h3[i] = tuple(h_edge)
            i += 1

H3 = hnx.Hypergraph(h3)
```
